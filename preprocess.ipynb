```python
from bs4 import BeautifulSoup
import os
```

```python
# Specify the input and output directories
input_directory = 'html_files'
output_directory = 'lecture_transcripts'

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate through HTML files in the input directory
for filename in os.listdir(input_directory):
    if filename.endswith(".html"):
        file_path = os.path.join(input_directory, filename)

        # Read the HTML content from the file
        with open(file_path, 'r', encoding='utf-8') as file:
            html_content = file.read()

        # Parse the HTML content with BeautifulSoup
        soup = BeautifulSoup(html_content, 'html.parser')

        # Create the output file path in the results directory
        output_file_path = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}_result.txt")

        tfile = open(output_file_path, 'a+')

        for i in soup.find_all('div', {'class': 'css-1shylkf'}):
            for p in i.find('div', {'class': 'phrases'}):
                print(p.text)

                # Write the extracted text to the output file
                tfile.write(p.text + "\n")

print("Extraction complete.")
```
